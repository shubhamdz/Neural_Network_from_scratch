### Coding of single and layer of neuron 

```python
inputs = [1,2,3]
weight = [0.2,0.8,-0.5]
bais = 2

output = (inputs[0]*weight[0]+inputs[1]*weight[1]+inputs[2]*weight[2] + bais)

print(output)
```

```python
#nueron with 4 input 

inputs = [1.0,2.0,3.0,2.5]
weight = [0.2,0.8,-0.5,1.0]
bais = 2

output = (inputs[0]*weight[0]+inputs[1]*weight[1]+inputs[2]*weight[2]+inputs[3]*weight[3] + bais)

print(output)
```

```python
#coding layer of neuron

inputs = [1.0,2.0,3.0,2.5]
weights1 = [0.2, 0.8, -0.5, 1]
weights2 = [0.5, -0.91, 0.26, -0.5]
weights3 = [-0.26, -0.27, 0.17, 0.87]

bias1 = 2
bias2 = 3
bias3 = 0.5

output = [inputs[0]*weights1[0]+ # output of first neuron
          inputs[1]*weights1[1]+
          inputs[2]*weights1[2]+
          inputs[3]*weights1[3]+
          bias1,
         
          inputs[0]*weights2[0]+ # output of second neuron
          inputs[1]*weights2[1]+
          inputs[2]*weights2[2]+
          inputs[3]*weights2[3]+
          bias2,
         
          inputs[0]*weights3[0]+ # output of third neuron
          inputs[1]*weights3[1]+
          inputs[2]*weights3[2]+
          inputs[3]*weights3[3]+
          bias3]

print(output)
```

```python
# same code with some code optimisation(using loops)

#coding layer of neuron

inputs = [1.0,2.0,3.0,2.5]
weights = [[0.2, 0.8, -0.5, 1],[0.5, -0.91, 0.26, -0.5],[-0.26, -0.27, 0.17, 0.87]]

bias = [2.0,3.0,0.5]

output = []

for i in range(len(weights)):
    neuron_weight = weights[i]
    neuron_bias = bias[i]
    output_neuron = neuron_bias
    
    for j in range(len(inputs)):
        output_neuron += inputs[j]*neuron_weight[j]
        
    output.append(output_neuron)
    
print(output)
```

```python
# same code with some code optimisation(using loops and using zip file)

#coding layer of neuron

inputs = [1.0,2.0,3.0,2.5]
weights = [[0.2, 0.8, -0.5, 1],[0.5, -0.91, 0.26, -0.5],[-0.26, -0.27, 0.17, 0.87]]

bias = [2.0,3.0,0.5]

output = []

for neuron_weight,neuron_bias in zip(weights,bias):  # this function will take first weight value and bias value and then second value and so on ...
    neuron_output = 0
    
    for n_input,weight in zip(inputs,neuron_weight):
        neuron_output += n_input*weight
    
    neuron_output += neuron_bias
    
    output.append(neuron_output)

print(output)
```

### coding numpy and dot product for layer neuron

```python
import numpy as np
```

```python
#codind dot product of single neuron using numpy
# np.dot(a,b) --> works as the row of a matrix(or vector) of "a" is multiplied with column of matrix(or vector) of "b"
a = [1,2,3]
b = [0.2,0.8,-0.5]

inputs = np.array(a)
weights = np.array(b)
bias = 2
print(inputs.shape,weights.shape)
output = np.dot(inputs,weights) + bias 
print(output)
```

"1‑D arrays in NumPy don’t carry explicit row or column vector labels". They’re simply sequences of numbers.

np.dot, when applied to 1‑D arrays, performs the inner product, which is mathematically equivalent to a matrix multiplication where one vector is treated as a row and the other as a column.

```python
#coding dot product of layer of neuron using numpy
a = [1.0,2.0,3.0,2.5]
b = [[0.2, 0.8, -0.5, 1],[0.5, -0.91, 0.26, -0.5],[-0.26, -0.27, 0.17, 0.87]]

inputs = np.array(a)
weights = np.array(b)
bias = [2.0,3.0,0.5]

print(inputs.shape,weights.shape)

output = np.dot(weights,inputs) + bias # here input is behaving as a column vector 
print(output)
```

```python
#coding dot product of layer of neuron using numpy
a = [1.0,2.0,3.0,2.5]
b = [[0.2, 0.8, -0.5, 1],[0.5, -0.91, 0.26, -0.5],[-0.26, -0.27, 0.17, 0.87]]

inputs = np.array(a)
weights = np.array(b)
bias = [2.0,3.0,0.5]

output = np.dot(inputs,weights.T) + bias #but here input is behaving as a row vector so, 1D array in NumPy does not carry explicit row or column vector label
print(output)
```

```python
#coding dot product of batch of data using numpy
a = [[1.0,2.0,3.0,2.5],[2.0,5.0,-1.0,2.0],[-1.5,2.7,3.3,-0.8]]
b = [[0.2, 0.8, -0.5, 1],[0.5, -0.91, 0.26, -0.5],[-0.26, -0.27, 0.17, 0.87]]

inputs = np.array(a)
weights = np.array(b)
bias = [2.0,3.0,0.5]

output = np.dot(inputs,weights.T) + bias
print(output)
```
